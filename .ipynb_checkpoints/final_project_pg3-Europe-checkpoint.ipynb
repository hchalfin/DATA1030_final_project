```python
gender = ['F' if name in female_led_countries else 'M' for name in covid_world['name']]
covid_world = covid_world.assign(leader_gender = gender)
covid_world.tail()
```

```python
today = pd.Timestamp(2020, 11, 29)
todays_data = covid_world.query('date == @today').reset_index(drop=True)
todays_data = todays_data.drop(['date'],axis=1)
Europe_str = 'Europe'
todays_data = todays_data.query('continent == @Europe_str').reset_index()
```

```python
todays_data.columns
```

```python
y = todays_data['leader_gender']
# X = todays_data.loc[:, 'new_cases_smoothed_per_million': 'tests_per_case']
# X = todays_data.loc[:, ['total_cases_per_million', 'total_deaths_per_million', 'total_tests_per_thousand']]
X = todays_data.loc[:, ['total_cases_per_million', 'total_deaths_per_million']]
Country = todays_data['name']
print(y)
print()
print(X)
print()
print(Country)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GroupShuffleSplit, GroupKFold, StratifiedKFold
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import xgboost
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import accuracy_score


max_depths = [1,3,10,30,100]

param_grid = {"learning_rate": [0.03],
              "n_estimators": [10000],
              "seed": [0],
              #"reg_alpha": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
              #"reg_lambda": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
              "missing": [np.nan], 
              #"max_depth": [1,3,10,30,100],
              "colsample_bytree": [0.9],              
              "subsample": [0.66]}

train_scores = []
val_scores = []
test_scores = []
```

```python
# init GroupShuffleSplit as follows
gss = GroupShuffleSplit(n_splits=3, train_size=.8, random_state=42)
for nth_split, (other_idx, test_idx) in enumerate(gss.split(X, y, Country)):    # group by Country
    # extract X_other, X_test, y_other, y_test, Country_other, Country_test
    X_other, y_other = X.iloc[other_idx], y.iloc[other_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]
    Country_other = Country.iloc[other_idx]
    Country_test = Country.iloc[test_idx]
    # init GroupKFold, 3 folds
    gkf = GroupKFold(n_splits=3)
    # create a temporary dictionary for kfold and test data
    random_split_tmp = {'kfold_train':[], 'kfold_cv':[], 'test': []}
    for train_idx, cv_idx in gkf.split(X_other, y_other, Country_other):    # group by Country_other
        # extract X_train, X_train, y_cv, y_cv, Country_train, Country_cv
        X_train, X_cv = X_other.iloc[train_idx], X_other.iloc[cv_idx]
        y_train, y_cv = y_other.iloc[train_idx], y_other.iloc[cv_idx]
        Country_train = Country_other.iloc[train_idx]
        Country_cv = Country_other.iloc[cv_idx]
        # check if ant Country exist in any of the two splits
        if set(Country_train.unique()) == set(Country_cv.unique()) or\
           set(Country_train.unique()) == set(Country_test.unique()) or\
           set(Country_cv.unique()) == set(Country_test.unique()):
            raise ValueError('Country exists in two or more of the splits')
        # init StandardScaler
        ss = StandardScaler()
        # fit_transform on train
        X_train = ss.fit_transform(X_train)
        # transform cv and test
        X_cv = ss.transform(X_cv)
        X_test = ss.transform(X_test)

        print('Countrys in train:',Country_train.unique())
        print('Countrys in CV:',Country_cv.unique())
        print('Countrys in test:',Country_test.unique())
        
        
        for d in max_depths:
            print('max_depth:', d)
            param_grid['max_depth'] = [d]
            XGB = xgboost.XGBClassifier()
            XGB.set_params(**ParameterGrid(param_grid)[0])
            XGB.fit(X_train,y_train,early_stopping_rounds=50,eval_set=[(X_cv, y_cv)], verbose=False)
            
            y_train_pred = XGB.predict(X_train)
            new_train_score = accuracy_score(y_train,y_train_pred)
            train_scores.append(new_train_score)
            print('the train accuracy score:', new_train_score)
            
            y_cv_pred = XGB.predict(X_cv)
            new_cv_score = accuracy_score(y_cv,y_cv_pred)
            val_scores.append(new_cv_score)
            print('the cv accuracy score:', new_cv_score)
            
            y_test_pred = XGB.predict(X_test)
            new_test_score = accuracy_score(y_test,y_test_pred)
            test_scores.append(new_test_score)
            print('the test accuracy score:', new_test_score)
            
            print()
```

```python
print('training scores:', train_scores)
print()
print('validation scores:', val_scores)
print()
print('test scores:', test_scores)
print()
```
